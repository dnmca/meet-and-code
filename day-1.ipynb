# Булева логіка 

Було б корисно мати змогу порівнювати значення у процесі їх використання. Для цього ми можемо використати вбудовані **оператори порівняння**.


| Оператор | результат | 
|----|---|
| x == y | True якщо x та y мають одне і те ж значення |
| x != y | True якщо x та y не мають те саме значення |
| x < y | True якщо x менше y |
| x > y | True якщо x більше y |
| x <= y | True якщо x менше-рівно y |
| x >= y | True якщо x більше-рівно y |

Зауважте, що ці оператори повертають булеве значення (`True` або `False`).

Відповідно, якщо якийсь з цих операторів не повертає `True`, то він відповідно повертає `False`.

Давайте спробуємо деякі з них.

````python
x = 5       # присвоюємо 5 змінній x
x == 5      # Перевіряємо чи x рівно 5
````

*Зауважте, що `==` це не те саме що `=`*

````python
x > 7
````

Чудово! Чи можемо ми тепер з'єднати декілька таких порівнянь? На щастя, Python дає нам змогу це зробити за допомогою **логічних операторів**:


| Оператор | Результат | 
|----|---|
| x or y | True якщо хоча б один з них True |
| x and y  | True тільки тоді якщо обидва True |
| not x | True тільки тоді якщо x є False |

Ось деякі приклади використання таких операторів:
```
True  and True  is True
True  and False is False
False and False is False

False or False is False
True  or True  is True
True  or False is True

not True  is False
not False is True
```

За домопогою цих операторів ми можемо об'єднувати декілька булівських операторів. 

Найпростішим прикладом використання буде перевірка того, чи число лежить у певному діапазоні.

````python
x = 14
# Перевірити чи x лежить у діапазоні 10..20

( x > 10 ) and ( x < 20)
````

Як бачите, дужки тут теж дуже допомагають, роблячи код більш читабельним. 

А що ж станеться ящок ми матимемо справу з якоюсь складною булівською логікою?


````python
# перевірити чи x є кратним 2
# перевірити чи x є кратним 3
# перевірити чи x не ділиться на 6 = 2*3.
x = 14

not (( x % 2 == 0 ) and ( x % 3 == 0))
````

З таким точно важко працювати!


Для того щоб зробити цей код більш зрозумілими, ми можемо використати проміжні змінні, як ми робили раніше:

````python
x = 14

xDivBy2 = ( x % 2 ) == 0 # перевірити чи x є кратним 2
xDivBy3 = ( x % 3 ) == 0 # перевірити чи x є кратним 3

not (xDivBy2 and xDivBy3)
````
